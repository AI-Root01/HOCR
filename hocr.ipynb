### 2 Reconstruir las tablas de los pdfs usando BeautifulSoup y Pandas

Primero veamos lo que es .hocr
HOCR es un estándar abierto para representar el análisis de maquetación de documentos y los resultados de OCR Reconocimiento Óptico de Caracteres (OCR) dentro de un archivo HTML normal.

Por ejemplo
- ocr_page	
    - La imagen completa.
- ocr_carea	
    - Un área o bloque de texto.
- ocr_line	
    - Una línea de texto completa.
- ocrx_word	
    - Una palabra individual.

Formato estándar:
```html
<span class="ocrx_word" title="bbox 10 20 100 40; x_wconf 95">Total:</span>

```
bbox x0 y0 x1 y1: Son las coordenadas de la caja delimitadora (Bounding Box) en píxeles.
- x0, y0: Esquina superior izquierda.
- x1, y1: Esquina inferior derecha.
x_wconf 95: Nivel de confianza del OCR al leer esa palabra (de 0 a 100).

Recursos: https://kba.github.io/hocr-spec/1.2/

# <p style="text-align: center;">Resultado final del script para el archivo 1C_5.hocr </p>

````python
import pandas as pd
import re
import os
from bs4 import BeautifulSoup
from config.rutas import HOCR_DIR, PDF_DIR, RESULTADOS_DIR

#Rutas de los archivos
HOCR_FILE=os.path.join(HOCR_DIR, "1C_5.hocr")
OUTPUT_CSV=os.path.join(RESULTADOS_DIR, "1C_5.csv")

#función para parsear el archivo HOCR y extraer texto con sus coordenadas
def parse_hocr_to_df(ruta_hocr):

    with open(ruta_hocr, 'r', encoding='utf-8') as f:
        soup = BeautifulSoup(f, 'html.parser')

    data = []
    # Buscamos todas las palabras (ocr_line o ocrx_word)
    for word in soup.find_all('span', class_='ocr_line'):
        text = word.get_text().strip()
        if not text: continue
        
        # Extraer bbox del atributo title: 'bbox 42 73 340 115; x_wconf 18'
        title = word.get('title', '')
        match = re.search(r'bbox (\d+) (\d+) (\d+) (\d+)', title)
        if match:
            x0, y0, x1, y1 = map(int, match.groups())
            data.append({
                'text': text,
                'x0': x0, 'y0': y0, 'x1': x1, 'y1': y1,
                'width': x1 - x0,
                'height': y1 - y0
            })

    return pd.DataFrame(data)

#función para reconstruir la tabla a partir del DataFrame de texto y coordenadas, usando las zonas definidas
def reconstruct_table(df, zones):

    df = df.sort_values(['y0', 'x0']).reset_index(drop=True) #para ordenar por coordenadas
    df['line'] = (df['y0'].diff() > 10).cumsum() # para agrupar por filas con una tolerancia de 10px para detectar cambios de línea
    
    structured_data = []
    
    for _, line_df in df.groupby('line'):
        row_dict = {zone['name']: [] for zone in zones}
        
        for _, row in line_df.iterrows():
            # Determinar a qué columna pertenece según su coordenada X
            centro_x = (row['x0'] + row['x1']) / 2
            for zone in zones:
                if zone['x_min'] <= centro_x <= zone['x_max']:
                    row_dict[zone['name']].append(row['text'])
                    break
        
        # Limpiar y unir texto
        final_row = {k: " ".join(v).strip() for k, v in row_dict.items()}
        if any(final_row.values()): # Evitar filas vacías
            structured_data.append(final_row)
            
    return pd.DataFrame(structured_data)


f = 4.16
COLUMNAS_BBVA = [
{"name": "Area_1", "x_min": 54 * f,  "x_max": 318 * f},
{"name": "Area_2", "x_min": 356 * f,  "x_max": 397 * f},
{"name": "Area_3", "x_min": 406 * f,  "x_max": 458 * f},
{"name": "Area_4", "x_min": 462 * f,  "x_max": 487 * f},
{"name": "Area_5", "x_min": 487 * f,  "x_max": 582 * f}
]

#Mapeamos el archivo HOCR a un DataFrame con texto y coordenadas
print("Leyendo el archivo HOCR y convirtiéndolo a DataFrame...")
df = parse_hocr_to_df(HOCR_FILE)
print("DataFrame creado con éxito. Mostrando las primeras filas:")
print(df.head(10))
print("--------------------------------------------------------------------")

df_coords = parse_hocr_to_df(HOCR_FILE)
df_final = reconstruct_table(df_coords, COLUMNAS_BBVA)
df_final.to_csv(OUTPUT_CSV, index=False, encoding='utf-8-sig') #por si hay caracteres especiales
print(f"Guardado en {OUTPUT_CSV}")


````

````python
import pandas as pd
import re
import os
import numpy as np
from bs4 import BeautifulSoup
from sklearn.cluster import KMeans
from config.rutas import HOCR_DIR, PDF_DIR, RESULTADOS_DIR

HOCR_FILE = os.path.join(HOCR_DIR, "1C_5.hocr")
OUTPUT_CSV = os.path.join(RESULTADOS_DIR, "1C_kmeans.csv")
N_COLUMNAS = 4 # El numero de areas que queremos detectar

def parse_hocr_to_df(ruta_hocr):
    with open(ruta_hocr, 'r', encoding='utf-8') as f:
        soup = BeautifulSoup(f, 'html.parser')

    data = []
    for word in soup.find_all('span', class_='ocrx_word'):
        text = word.get_text().strip()
        if not text: continue
        
        title = word.get('title', '')
        match = re.search(r'bbox (\d+) (\d+) (\d+) (\d+)', title)
        if match:
            x0, y0, x1, y1 = map(int, match.groups())
            data.append({
                'text': text,
                'x0': x0, 'y0': y0, 'x1': x1, 'y1': y1,
                'cx': (x0 + x1) / 2 # Centroide X para el modelo
            })

    return pd.DataFrame(data)

def get_kmeans_zones(df, n_clusters): #para calcular zonas dinámicas con K-Means

    X = df[['cx']].values # Usamos solo el centroide X para clustering horizontal
    kmeans = KMeans(n_clusters=n_clusters, random_state=1, n_init='auto') 
    df['cluster'] = kmeans.fit_predict(X)
    
    # Ordenamos clusters de izquierda a derecha
    centers = df.groupby('cluster')['cx'].mean().sort_values()
    
    zones = []
    for i, (cluster_id, center) in enumerate(centers.items()):
        cluster_data = df[df['cluster'] == cluster_id]
        zones.append({
            "name": f"Area_{i+1}",
            "x_min": cluster_data['x0'].min(),
            "x_max": cluster_data['x1'].max()
        })
    return zones

def reconstruct_table(df, zones): # para reconstruir la tabla a partir del DataFrame de texto y coordenadas, usando las zonas definidas
    df = df.sort_values(['y0', 'x0']).reset_index(drop=True) 
    df['line'] = (df['y0'].diff() > 25).cumsum() 
    
    structured_data = []
    for _, line_df in df.groupby('line'):
        row_dict = {zone['name']: [] for zone in zones}
        
        for _, row in line_df.iterrows():
            centro_x = (row['x0'] + row['x1']) / 2
            for zone in zones:
                # Usamos las zonas calculadas por el modelo
                if zone['x_min'] <= centro_x <= zone['x_max']:
                    row_dict[zone['name']].append(row['text'])
                    break

        final_row = {k: " ".join(v).strip() for k, v in row_dict.items()}
        if any(final_row.values()): 
            structured_data.append(final_row)
            
    return pd.DataFrame(structured_data)

print("Leyendo el archivo HOCR")
df_coords = parse_hocr_to_df(HOCR_FILE)

print("Calculando zonas dinámicas con K-Means")
zonas_dinamicas = get_kmeans_zones(df_coords, N_COLUMNAS)

print("Reconstruyendo tabla con zonas calculadas")
df_final = reconstruct_table(df_coords, zonas_dinamicas)

df_final.to_csv(OUTPUT_CSV, index=False, encoding='utf-8-sig')
print(f"Guardado exitosamente en {OUTPUT_CSV}")
````
